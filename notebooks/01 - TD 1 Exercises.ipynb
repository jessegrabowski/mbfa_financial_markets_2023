# Section 1

```python
# State-based returns for each asset
r_A = pd.Series(np.array([0.3,   0.1, -0.1]), index=['Bull', 'Bear', 'Crisis'])
r_B = pd.Series(np.array([0.25, -0.15]), index=['Bull', 'Crisis'])

# Joint probabilities of each state
P_AB = pd.DataFrame(np.array([[0.10, 0.10, 0.4],
                              [0.15, 0.15, 0.1]]),
                    index=pd.MultiIndex.from_product([['B'], ['Bull', 'Crisis']]),
                    columns=pd.MultiIndex.from_product([['A'], ['Crisis', 'Bear', 'Bull']]))
```

```python
P_AB
```

## Q1: Expected Correlation?

Should be positive, because the probability that both assets are in the bull state or crisis state is higher than the probability they are in opposite states.

## Q2: Marginal probabilties?

Sum across the rows/columns of the probability distribution. Verify the result is a valid probability distribution by checking the resulting array adds up to 1

```python
P_A = P_AB.sum(axis=0).droplevel(0)
P_A
```

```python
assert P_A.sum() == 1
```

```python
P_B = P_AB.sum(axis=1).droplevel(0)
P_B
```

```python
assert P_B.sum() == 1
```

## Q3: Expected returns?

Need to compute $$\mathbb E[r(x)] = \sum_{z}p(r = z)r(x, z)$$


For $x \in \{A, B\}$ and $z \in \{\text{Bull}, \text{Bear}, \text{Crisis}\}$

```python
μ_A = (P_A * r_A).sum()
μ_A
```

```python
μ_B = (P_B * r_B).sum()
μ_B
```

## Q4: Std and Variance?

Compute variance as:
$$\text{Var}[r(x)] = \sum_{z}p(r = z)\left ( r(x, z) - \mathbb E [r(x)] \right )^2$$

```python
σ2_A = (P_A * (r_A - μ_A) ** 2).sum()
σ_A = np.sqrt(σ2_A)

print(f'Variance of A: {σ2_A:0.5}')
print(f'Std of A: {σ_A:0.5}')
```

```python
σ2_B = (P_B * (r_B - μ_B) ** 2).sum()
σ_B = np.sqrt(σ2_B)

print(f'Variance of B: {σ2_B:0.5}')
print(f'Std of B: {σ_B:0.5}')
```

## Q5: Covariance?

```python
# Make all combinations of r_A * r_B using broadcasting
r_AB = r_A.values[::-1] * r_B.values[:, None]
r_AB
```

```python
σ_AB = (P_AB * r_AB).sum().sum() - μ_A * μ_B
σ_AB
```

## Q6: Correlation

```python
σ_AB / σ_A / σ_B
```

## Conditional probabilities of A given B = Bull

```python
P_A_given_Bull = (P_AB.loc[('B', 'Bull')] / P_B['Bull']).droplevel(0)
P_A_given_Bull
```

## Conditional mean of  A given B = Bull

```python
μ_A_given_Bull = (r_A * P_A_given_Bull).sum()
μ_A_given_Bull
```

## Conditional variance of A give B = Bull

```python
(P_A_given_Bull * (r_A - μ_A_given_Bull) ** 2).sum()
```

## Q1: Calculate mean and variance

```python
μ_A = (r_A * P_A).sum()
μ_B = (r_B * P_B).sum()
print(f'Mean of asset A: {μ_A:0.5f}')
print(f'Mean of asset B: {μ_B:0.5f}')
```

```python
σ2_A = (r_A ** 2 * P_A).sum() - μ_A ** 2
σ2_B = (r_B ** 2 * P_B).sum() - μ_B ** 2
print(f'Variance of asset A: {σ2_A:0.5f}')
print(f'Variance of asset B: {σ2_B:0.5f}')
```

## Q2: Mean-Variance Utility

Given a utility function $$U(r) = \mu_r - \frac{A}{2}\sigma^2_r$$

 Which asset do you prefer?

```python
from functools import partial

def utility(A, μ, σ2):
    return μ - A / 2 * σ2
```

```python
UA = partial(utility, μ=μ_A, σ2=σ2_A)
UB = partial(utility, μ=μ_B, σ2=σ2_B)
```

```python
A0 = (μ_B - μ_A) * (σ2_B / 2 - σ2_A / 2) ** -1
```

```python
A_grid = np.linspace(A0 - 2*A0, A0 + 2*A0, 100)
fig, ax = plt.subplots()
ax.plot(A_grid, UA(A_grid))
ax.plot(A_grid, UB(A_grid))
ax.scatter(A0, UA(A0))
plt.show()
```

## Higher Order Moments

```python
Sk_A = (P_A * ((r_A - μ_A) / np.sqrt(σ2_A)) ** 3).sum()
Sk_B = (P_B * ((r_B - μ_B) / np.sqrt(σ2_B)) ** 3).sum()
print(f'Skew of asset A: {Sk_A:0.5f}')
print(f'Skew of asset B: {Sk_B:0.5f}')
```

```python
Kr_A = (P_A * ((r_A - μ_A) / np.sqrt(σ2_A)) ** 4).sum()
Kr_B = (P_B * ((r_B - μ_B) / np.sqrt(σ2_B)) ** 4).sum()
print(f'Kurtosis of asset A: {Kr_A:0.5f}')
print(f'Kurtosis of asset B: {Kr_B:0.5f}')
```

### Skew Visualized

```python
from scipy import stats
x_grid = np.linspace(-3, 3, 100)
fig, ax = plt.subplots(1, 2, figsize=(14,4))
for a in [0.1, 0.25, 0.5, 1, 2, 5]:
    d = stats.skewnorm(a=a, loc=0, scale=1)
    ax[0].plot(x_grid, d.pdf(x_grid), label=f'a = {a:0.2f}, Sk = {d.stats("s"):0.2f}')
ax[0].legend()
for a in [0.1, 0.25, 0.5, 1, 2, 5]:
    d = stats.skewnorm(a=-a, loc=0, scale=1)
    ax[1].plot(x_grid, d.pdf(x_grid), label=f'a = {-a:0.2f}, Sk = {d.stats("s"):0.2f}')
ax[1].legend()
plt.show()
```

### Kurtosis Visualized

```python
x_grid = np.linspace(-3, 3, 100)
fig, ax = plt.subplots()
nu_values = [1, 4.1, 5, 7, 10, 100]
for nu in reversed(nu_values):
    d = stats.t(nu, loc=0, scale=1)
    ax.plot(x_grid, d.pdf(x_grid), label=f'nu = {nu:0.2f}, Kr = {d.stats("k"):0.2f}')
ax.legend()
```

In general, investors should like positive skewness, because it is "upside-weighted risk". They should not like kurtosis, because it's like super variance.

# Short Selling

## Initial collateral

You manage a Long-Short Fund. You wish to short 100 shares of the company Keep Quiet at 73€ by stocks. The initial margin is 55%, how much of collateral do you have to invest?

```python
M0 = 0.55
P0 = 73
Q = 100

V0 = P0 * Q
C = M0 * V0
C
```

## Stock falls

Assume now, that the stock price comes down to 65€. What is your current margin? Considering that the maintenance margin is 30%, what is happening?

```python
P1 = 65
V1 = P1 * Q
price_change = P0 / P1 - 1
M1 = price_change + C / V1
M1
```

Our margin has increased from 55% to 62%, and we're further from being margin called. Our short position is paying off!

## Stock rises

Assume then, that the stock price rises to 80€. What is then your current margin?

```python
P2 = 80
V2 = P2 * Q
price_change = P0 / P2 - 1
M2 = price_change + C / V2
M2
```

The margin has now fallen to 50%. We are closer to being margin called. Our short position is losing money.

## Margin Call

At which price do you have a margin call?

```python
M_C = 0.3
P_call = (Q * P0 + C) / (M_C * Q + Q)
P_call
```

# S1 - Indices

```python
prices = pd.DataFrame(np.array([[20, 12, 90],
                                [22, 15, 82],
                                [22, 15, 41]], dtype='float64'),
                      index=[0, 1, 2],
                      columns=['A', 'B', 'C'])
quantities = pd.DataFrame(np.array([[200, 400, 100],
                                    [200, 400, 100],
                                    [200, 400, 200]], dtype='float64'),
                          index=[0, 1, 2],
                         columns=['A', 'B', 'C'])
```

## Calculate two indices

### Price-weighted index

```python
price_changes = prices.apply(np.log).diff(axis=0)
quantity_changes = quantities.apply(np.log).diff(axis=0)
splits = price_changes.apply(np.exp).pow(-1) * ((price_changes + quantity_changes).abs() < 1e-8)
splits = splits.fillna(1).replace({0:1})
```

```python
price_weights = pd.DataFrame(np.full_like(prices, 1 / prices.shape[1]), index=prices.index, columns=prices.columns)
price_index = (prices * price_weights * splits).sum(axis=1)
price_index
```

```python
price_index_normalized = price_index / price_index.loc[0] * 100
price_index_normalized
```

### Capitalization-weighted index

```python
mkt_cap_weights = (prices * quantities)
mkt_cap_index = mkt_cap_weights.sum(axis=1)
mkt_cap_index
```

```python
mkt_cap_index_normalized = mkt_cap_index / mkt_cap_index.loc[0] * 100
mkt_cap_index_normalized
```

```python
weights = pd.concat([(price_weights * splits).unstack(), mkt_cap_weights.apply(lambda x: x / x.sum(), axis=1).unstack()], axis=1).rename(columns = {0:'Price', 1:'Market Cap'})
indices = pd.concat([price_index, mkt_cap_index], axis=1).rename(columns = {0:'Price', 1:'Market Cap'})
```

```python
fig, ax = plt.subplots(3, 1, figsize=(12, 6), sharex=True)
for i, axis in enumerate(fig.axes):
    weights.xs(level=1, key=i).plot.bar(ax=axis, legend=False)
    axis.set_title(f'Period {i} Weights')
ax[0].legend()
fig.tight_layout()
plt.show()
```

```python
from matplotlib.patches import Patch
fig, ax = plt.subplots()
indices['Price'].plot(ax=ax, marker='o', label='Price Index')
indices['Market Cap'].plot(ax=ax, marker='o', label='Capitalization Index')
ax.legend()
plt.show()
```

# S2 - Buying on Margin

## Question 1
You manage a SRI fund and you want to buy 100 B shares at 15€ per share. The security deposit is 60%, how much of your cash should you mobilize for this purchase?

$$\text{deposit} = PQd$$

Where $P$ is the price of the shares, $Q$ is the quantity of shares, and $d \in [0, 1]$ is the deposit requirement.

```python
Q = 100
P = 15
V1 = Q * P
deposit_pct = 0.6
deposit = V1 * deposit_pct
deposit
```

We can also compute the size of the loan we recieve:

$$ \text{loan} = PQ(1 - d) = PQ - \text{deposit}$$

```python
loan = V1 - deposit
loan
```

## Question 2
Share B price drops to 11€. What is now your margin? If the maintenance margin is 40%, what happens? Build the account associated with these transactions.

The margin is the percent change in the value of the position __relative to the loan__. Call $P^\prime$ the new price, then:

$$M = \frac{P^\prime Q - \text{loan}}{P^\prime Q} = \frac{P^\prime Q - (1 - d)PQ}{P^\prime Q}$$

```python
P2 = 11
V2 = Q * P2
price_change_pct = P / P2 - 1
M = (V2 - loan) / V2
M
```

## Question 3

If the margin call happens at 40% margin, at what price will we get margin called?

Solve the $M$ formula for $P^\prime$:

$$\begin{align}
M &= \frac{P^\prime Q - (1 - d)PQ}{P^\prime Q} \\
P^\prime Q M &= P^\prime Q - (1 - d)P Q \\
P^\prime M - P^\prime &= -(1 - d)P \\
P^\prime &= \frac{(1 - d) P}{(1 - M)}
\end{align}$$ 

```python
margin_call = 0.4
P_prime = (1 - deposit_pct) * P / (1 - margin_call)
P_prime
```
